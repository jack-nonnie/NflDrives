```python
from bs4 import BeautifulSoup
import urllib
import re
import time
import pandas as pd
import os
import numpy as np
from urllib.parse import urlparse
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#from 401326308 to 401326492
espn = "https://www.espn.com/nfl/team/schedule/_/name/gb/season/2021"
espn_content = urllib.request.urlopen(espn)
espn_html = BeautifulSoup(espn_content, "html.parser")
espn_html
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(ChromeDriverManager().install(), options = options)
driver.get(espn)

```

```python
arr = []
#https://www.espn.com/nfl/playbyplay/_/gameId/401326319
for game in range(401326308, 401326492):
    driver.get("https://www.espn.com/nfl/playbyplay/_/gameId/" + str(game))
    print(game)
    lst = driver.find_elements_by_class_name("accordion-header")
    drives = driver.find_elements_by_class_name("drive-list")
    for i in range(len(lst)):
        line = []
        line.append(lst[i].find_element_by_class_name("home-logo").find_element_by_tag_name("img").get_attribute("src").split("/")[9].split(".")[0])
        line.append(lst[i].find_element_by_class_name("headline").text)
        line.append(lst[i].find_element_by_class_name("drive-details").text)
        if(drives[i].find_elements_by_tag_name("h3")[0].get_attribute("textContent") != "" or len(drives[i].find_elements_by_tag_name("h3")) == 1):
            line.append(drives[i].find_elements_by_tag_name("h3")[0].get_attribute("textContent"))
        else:
            line.append(drives[i].find_elements_by_tag_name("h3")[1].get_attribute("textContent"))
        line.append(game)
        arr.append(line)
print(arr)
          

          
```

```python
#print(drives[10].find_elements_by_tag_name("h3")[1].get_attribute("textContent"))
#print(drives[10].find_elements_by_tag_name("h3"))
for line in arr:
    if(line[3] != ""):
        if(line[3][0] != "1"):
            print(line[3])
drivesData = np.array(arr)
df = pd.DataFrame(drivesData, columns = ["team", "result", "details", "drive_start", "game_id"])

print(df["team"].unique())
print(df["result"].unique())

df.to_csv("drive_results.csv")
df.head()
```
